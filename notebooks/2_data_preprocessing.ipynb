```python


def extract_top_cast(cast_data):
    if isinstance(cast_data, str):  # If it's a string, convert it
        try:
            cast_data = ast.literal_eval(cast_data)  # Convert to list
        except (ValueError, SyntaxError):
            return []  # Return empty if conversion fails

    if isinstance(cast_data, list):  # Ensure it's now a list
        return [i.get('name', '') for i in cast_data[:3]]  # Extract top 3 names

    return []  # Return empty if not a valid list
```

```python
def extract_director(crew_data):
    if isinstance(crew_data, str):
        try:
            crew_data = ast.literal_eval(crew_data)
        except(ValueError, SyntaxError):
            return ""
        
    if isinstance(crew_data, list):
        for i in crew_data:
            if i.get("job") == "Director":
                return i.get('name', '')
    return ""
```
