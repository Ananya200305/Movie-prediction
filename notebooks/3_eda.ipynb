```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
merged_df = pd.read_csv('/Users/ananyaaggarwal/Desktop/Movie Prediction Project/data/cleaned_and_merged_data.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='budget', y='revenue', data=merged_df)
plt.title('Budget vs. Revenue')
plt.savefig('/Users/ananyaaggarwal/Desktop/Movie Prediction Project/reports/visualizations/budget_vs_revenue.png')
plt.show()
```

```python
top_directors = merged_df.groupby('director')['revenue'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.index, y=top_directors.values)
plt.xticks(rotation=90)
plt.title('Top 10 Directors by Average Revenue')
plt.savefig('/Users/ananyaaggarwal/Desktop/Movie Prediction Project/reports/visualizations/top_directors.png')
plt.show()
```

```python
# Step 1: Explode 'cast' column to separate each actor into individual rows
exploded_df = merged_df.explode('cast')

# Step 2: Group by individual actor and compute their average revenue
actor_revenue = exploded_df.groupby('cast')['revenue'].mean().sort_values(ascending=False)

# Step 3: Display the top 10 actors by average revenue
print(actor_revenue.head(10))  # Print the top 10 actors

# Step 4: Plot the top 10 actors
plt.figure(figsize=(12, 6))
sns.barplot(x=actor_revenue.head(10).index, y=actor_revenue.head(10).values, palette='viridis')

# Step 5: Format the plot
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.xlabel('Actor', fontsize=14)
plt.ylabel('Average Revenue ($)', fontsize=14)
plt.title('Top 10 Actors by Average Revenue', fontsize=16, fontweight='bold')

# Step 6: Save and Show
plt.savefig('/Users/ananyaaggarwal/Desktop/Movie Prediction Project/reports/visualizations/top_actors.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(10, 6))

# Plot histogram with log scale for better visualization
sns.histplot(merged_df['success_ratio'], bins=50, kde=True, color='royalblue')

# Formatting
plt.xlim(0, merged_df['success_ratio'].quantile(0.95))  # Exclude extreme outliers
plt.xlabel('Success Ratio (Revenue / Budget)', fontsize=14)
plt.ylabel('Number of Movies', fontsize=14)
plt.title('Distribution of Movie Success Ratios', fontsize=16, fontweight='bold')

# Save and Show
plt.savefig('/Users/ananyaaggarwal/Desktop/Movie Prediction Project/reports/visualizations/success_ratio_histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np

# Replace 0 budgets with NaN to avoid division errors
merged_df['success_ratio'] = merged_df['revenue'] / merged_df['budget'].replace(0, np.nan)

# Replace NaN values (optional)
merged_df['success_ratio'].fillna(0, inplace=True)

# Check if the column exists and is correctly calculated
print(merged_df['success_ratio'].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(10, 6))

# Plot histogram with log scale for better visualization
sns.histplot(merged_df['success_ratio'], bins=50, kde=True, color='royalblue')

# Formatting
plt.xlim(0, merged_df['success_ratio'].quantile(0.95))  # Exclude extreme outliers
plt.xlabel('Success Ratio (Revenue / Budget)', fontsize=14)
plt.ylabel('Number of Movies', fontsize=14)
plt.title('Distribution of Movie Success Ratios', fontsize=16, fontweight='bold')

# Save and Show
plt.savefig('/Users/ananyaaggarwal/Desktop/Movie Prediction Project/reports/visualizations/success_ratio_histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```
